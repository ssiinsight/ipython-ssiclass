##1.2 왜 데이터 분석을 위한 파이썬인가?

- 스터디원들이 공통적으로 말하는 것은 **'전처리와 후처리는 파이썬, 분석은 R'** 이라고 한다.
- 그 이유는 전처리와 후처리 하는데 필요한 것들이 파이썬으로는 정말 쉽게 할 수 있기 때문이다. 반면에 R로 하려면 굉장히 힘들다.
- 분석은 R로 하라는 이유는? R의 철학을 살펴보면 좋다.
- R은 통계학자가 만든 언어라 통계에 필요한 라이브러리들이 굉장히 많다. 컴퓨터 전공인 사람이 만들지 않아 컴공 전공인 사람들이 약간 의아한 부분이 있다. 나도 처음에 배울때 왜 이걸 이렇게 쓰지? 아직도 apply 함수가 잘 이해 되지 않는다. 스터디원들 말로는 C++이나 자바에도 이러한 개념이 있다고 하던데. 난 잘 모르겠다.
- R은 패키지를 만들었다고 해서 마음대로 배포할 수 없다. R에 라이브러리를 올리고 커미터가 확인을 해주어야 비로소 배포가 가능하다. 그래서 패키지들이 파이썬보다 신뢰가 간다. 그만큼 커미터들의 힘이 강력하다. 힘이 있는 곳에는 당연히 따라오는게 있지..말 안해도 알거라고 생각하고 스킵~
- 물론 유명한 파이썬 라이브러리들도 여러 사람들이 검증을 해주었기 때문에 신뢰가 간다. 파이썬 라이브러리가 신뢰가 가지 않는다는 소리가 아니니 참고

### 1.2.3 파이썬을 사용하면 안 되는 경우

- 파이썬은 [인터프리터 언어](http://seodh007.tistory.com/2)이기 때문에 Java나 C++ 같은 컴파일 언어보다 많이 느리다.
- 하지만 **개발자의 시간 비용은 CPU의 시간 비용보다 비싸**므로 대개는 이런 등가교환에 만족해한다. 그러나 실시간 거래 시스템처럼 매우 짧은 응답 시간을 필요로 하는 애플리케이션에서는 가능한 한 최고의 성능을 내고자 생산성은 떨어지지만 C++ 같은 저수준의 언어로 개발
- 파이썬은 동시다발적인 멀티쓰레드를 처리하거나 CPU에 집중된 많은 쓰레드를 처리하는 애플리케이션에 적합하지 않다. 이는 바로 [GIL(Global Interpreter Lock)](http://mcchae.egloos.com/11099578) 때문. 이 매커니즘은 인터프리터가 한 번에 하나의 파이썬 바이트 코드 명령만 실행하도록 한다. 그런데 대체로 빅데이터 처리 애플리케이션에서는 단일 클러스터가 적절한 시간 안에 데이터를 처리해야만 하므로 단일 프로세스, 멀티스레드 시스템을 선호

## 1.3 필수 파이썬 라이브러리

### 1.3.1 NumPy

- [Numpy](http://numpy.org)(Numerical Python의 줄임말), 과학계산용 파운데이션 패키지
- 빠르고 효율적인 다차원 배열 객체 ndarray
- 배열 원소를 다루거나 배열 간의 수학 계산을 수행하는 함수
- 디스크로부터 배열 기반의 데이터를 읽거나 쓸 수 있는 도구
- 선형대수 계산, 푸리에 변환, 난수 발생기
- 파이썬과 C, C++ 그리고 포트란 코드 통합 도구
- 데이터분석에서는 알고리즘에 사용할 데이터 컨테이너의 역할
- NumPy 배열은 파이썬 기본 자료 구조보다 훨씬 효율적인 방법으로 데이터를 저장하고 다룬다.
- [Matching Performance using Python list's, ndarray's, and pandas](http://www.benjaminmgross.com/matching-performance-using-python-listss-ndarrays-and-pandas/)

### 1.3.2 pandas

- pandas는 구조화된 데이터를 빠르고 쉬우면서도 다양한 형식으로 가공할 수 있는 풍부한 자료 구조와 함수 제공
- pandas는 파이썬을 강력하고 생상적인 데이터 분석 환경으로 만드는 데 꼭 필요(**이 책 저자가 pandas 개발자라는게 함정!**)
- 이 책에서는 주로 pandas의 주요 객체인 DataFrame을 다룰 텐데, 이 객체는 2차원 표 또는 행과 열을 나타내는 자료 구조(엑셀로 보면 이해하기 쉽다)
- pandas는 NumPy의 고성능 배열 계산 기능과 스프레드시트, SQL같은 관계형 데이터베이스의 유연한 데이터 조작 기능 조합. 세련된 인덱싱 기능으로 쉽게 데이터 재배치하고 잘게 조각내거나 집계하고 부분집합을 구할 수 있음
- 금융 데이터를 분석하려는 사용자에게 pandas는 풍부한 고성능 시계열 처리 기능과 금융 데이터에 딱 맞는 도구 제공. 사실 저자는 처음부터 금융 데이터 분석 애플리케이션용으로 pandas 설계
- pandas 개발자가 궁극의 데이터 분석 툴을 개발중이라는 스터디원의 제보. 상용으로 제공 가능성 매우 높음
- R의 data.frame 객체에서 따온 DataFrame. 이렇게 컴퓨터 세상에서는 좋은건 가져다 씀.
- pandas는 다차원으로 구조화 된 데이터를 뜻하는 계량 경제학 용어인 PANel DAta와 파이썬 데이터 분석Python data analysiS에서 따온 이름

### 1.3.4 IPython

- 표준 과학계산용 파이썬 도구 모음에 포함된 컴포넌트
- Interactive하고 강력한 생상적인 환경 제공
- IPython은 파이썬 코드를 작성하고 테스트하고 디버깅 할 수 있는 향상된 파이썬 쉘 제공
- 특히 데이터를 처리하고 matplotlib으로 데이터를 시각화 하는데 매우 유용
- IPython은 실행, 디버깅, 테스트 같은 파이썬을 필요로 하는 작업을 대부분 수행
- IPython을 웹브라우저와 연결할 수 있는 Mathematica 스타일의 HTML **노트북** 기능
- 그래프를 즉시 그려보거나 여러 줄을 편집할 수 있는 기능
- 문법 갖오 기증을 가진 Qt 프레임워크 기반의 GUI 콘솔
- 병렬 분산 컴퓨팅을 위한 기반 구조

### 1.3.5 SciPy

- [Scipy](http://scipy.org)는 과학계산 컴퓨팅 영역의 여러 기본 문제를 다루는 패키지 모음

- scipy.integrate: 수치적분 루틴과 미분방정식 해법기
- scipy.linalg: numpy.linalg에서 제공하는 것보다 더 확장된 선형대수 루틴과 매트릭스 분해
- scipy.optimize: 함수 최적화기와 방정식의 근을 구하는 알고리즘
- scipy.signal: 시그널 프로세싱 도구
- scipy.sparse: 희소 행렬과 희소 선형 시스템 풀이법
- scipy.special: 감마 함수처럼 흔히 사용되는 수학 함수를 구현한 포트란 라이브러리인 SPECFUN 확장
- scipy.stats: 표준 연속/이산 확률 분포(집적도 함수, 샘플러, 연속 분포 함수)와 다양한 통계 테스트, 그리고 좀 더 기술적인 통계 도구
- scipy.weave: 배열 계산을 빠르게 하기 위해 인라인 C++ 코드를 사용하는 도구

- NumPy와 Scipy를 함께 사용하면 확장 애드온을 포함한 MATLAB을 완벽하게 대체

### 1.6.3 import 컨벤션

- 파이썬 커뮤니티는 자주 사용되는 몇 가지 모듈에 대해 다음과 같은 네이밍 컨벤션 사용
- numpy는 np
- pandas는 pd
- matplotlib 모듈의 pyplot은 plt로 축약한 컨벤션

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- 이렇게 하면 np.arange는 NumPy에 있는 arrange 함수를 참조한다는 듯
- NumPy 같은 거대한 패키지에서 모든 것을 import하는 것(from numpy import *)은 파이썬 소프트웨어 개발에 있어 나쁜 사례
- 왜냐하면 큰 식당에 들어가서 먹지도 않을 각종 음식을 시키면 나머지를 준비하는 시간도 들어가고 그것을 내놓을 식탁도 모자랄테고 모든 것들이 부족하게 된다. 그러므로 꼭 필요한 것만 import 하는 습관을 갖도록 하자
